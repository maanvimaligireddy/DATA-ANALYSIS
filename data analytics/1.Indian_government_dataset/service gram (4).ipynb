DATASET DESCRIPTION:
The Bharatnet Service Ready Gram Panchayats dataset contains information about the readiness of Gram Panchayats (GPs) for BharatNet services across various states and districts in India. It consists of 716 entries with the following columns:

State: The name of the state or union territory.
District: The district within the state.
Total Gram Panchayats: The total number of GPs in the district.
Service Ready Gram Panchayats: The number of GPs that are ready for BharatNet services.
Key Observations:

 There are no missing values in the State, Total Gram Panchayats, or Service Ready Gram Panchayats columns.
One record has a missing district name, which may require further review.
The data provides a clear overview of service readiness by district, making it useful for evaluating connectivity progress and planning further expansions.

```python
import pandas as pd
import seaborn as sns

# Load the CSV file (ensure correct encoding)
df = pd.read_csv("BharatnetServiceReadyGPs31012024.csv", encoding="utf-8")

# Strip spaces and convert column names to lowercase for consistency
df.columns = df.columns.str.strip()

# Print columns to verify column names
print(df.columns)

# Check if 'State' exists, otherwise find the correct column name
if 'State' not in df.columns:
    print("Column 'State' not found! Available columns:", df.columns)

# Group by 'State' and sum 'Service Ready Gram Panchayats' (check column name)
grouped_data = df.groupby('State')['Service Ready Gram Panchayats'].sum()

# Generate colors for plotting
colors = sns.color_palette("viridis", len(grouped_data))

# Print the grouped data to verify
print(grouped_data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Strip spaces from column names (if not done earlier)
df.columns = df.columns.str.strip()

# Group by 'State' and sum 'Service Ready Gram Panchayats'
state_data = df.groupby('State')['Service Ready Gram Panchayats'].sum()

# Generate colors for plotting
colors = sns.color_palette("viridis", len(state_data))

# *Pie Chart*
plt.figure(figsize=(10, 8))
plt.pie(state_data, labels=state_data.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.axis('equal')  # Ensures pie chart is circular
plt.title("Service Ready Gram Panchayats by State")
plt.show()
```

# Distribution of Service Ready Gram Panchayats by State Category


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#
df = pd.read_csv("BharatnetServiceReadyGPs31012024.csv", encoding="utf-8")
import warnings
warnings.filterwarnings("ignore")


# Step 1: Print available columns to debug
print("Available Columns:", df.columns.tolist())

# Step 2: Clean column names (remove spaces, convert to lowercase)
df.columns = df.columns.str.strip().str.lower()
print("Cleaned Columns:", df.columns.tolist())  # Print again to verify

# Step 3: Check for 'state' column and rename if necessary
expected_column = "state"
if expected_column not in df.columns:
    possible_match = [col for col in df.columns if "state" in col.lower()]
    if possible_match:
        df.rename(columns={possible_match[0]: expected_column}, inplace=True)
        print(f"Renamed column '{possible_match[0]}' to '{expected_column}'")
    else:
        print("Error: No 'State' column found!")
        exit()  # Stop execution if no match is found

# Step 4: Group data and visualize
state_counts = df[expected_column].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=state_counts.index, y=state_counts.values, palette="viridis")
plt.xticks(rotation=45)
plt.xlabel("State")
plt.ylabel("Number of Service Ready Gram Panchayats")
plt.title("Top States with Highest Service Ready Gram Panchayats")
plt.show()
```
